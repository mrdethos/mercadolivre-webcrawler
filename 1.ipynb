```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = webdriver.Chrome()
driver.get('https://www.mercadolivre.com.br/ofertas')

# nome ✅, valor atual✅, valor antigo, valor total, desconto ✅, link da imagem ✅, prazo de frete, nome da loja vendedora ✅
# exportar para um json

# Extrai todos os anúncios da página atual
anuncios_container = driver.find_element_by_css_selector('.items_container')
```

```python
anuncios = anuncios_container.find_elements_by_css_selector('.promotion-item')

for anuncio in anuncios:
    
    nome_produto = anuncio.find_element_by_css_selector('.promotion-item__title').text
    
    preco_anterior = anuncio.find_element_by_css_selector('.promotion-item__price').text
    preco_produto = anuncio.find_element_by_css_selector('.promotion-item__price').text
    
    
    valor_atual = anuncio.find_element_by_css_selector('.andes-money-amount__fraction').text
    # Caso o vendedor seja encontrado, ele é adicionado, caso o contrário, o valor será "N/A"
    vendedor_elementos = anuncio.find_elements_by_css_selector('.promotion-item__seller')
    if len(vendedor_elementos) > 0:
        vendedor = vendedor_elementos[0].text
    else:
        vendedor = "N/A"
    
    # Para evitar adicionar o link da imagem codificada como base64, rola a opaguba até o trecho do anúncio, 
    # fazendo com que a url inteira carregue
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    actions = ActionChains(driver)
    actions.move_to_element(anuncio)
    actions.perform()
    link_imagem = WebDriverWait(anuncio, 20).until(EC.visibility_of_element_located((By.CSS_SELECTOR, '.promotion-item__img'))).get_attribute("src")
    
    
    #print("Nome:", nome_produto)
    #print("Vendedor:", vendedor)
    #print("Preço:", preco_produto)
    #print("Valor atual:", valor_atual)
    
    #print("Desconto:", desconto_produto)
    #print('Link da imagem:', link_imagem)
    print("--------------------")
```
